```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({
    "text.usetex": False,
    "font.family": "Arial",
})
```

```python
def bootstrap_gradient_variance(list_of_grads, repetitions: int = 50):
    all_grads = []

    for rep in range(repetitions):
        possible_choices = np.shape(list_of_grads)[1]
        indices = np.random.choice(possible_choices, size=possible_choices)
        samples_grads = list_of_grads[:, indices]
        gradient_variances = np.std(samples_grads, axis=1)**2
        all_grads.append(gradient_variances)

    return np.median(all_grads, axis=0), np.percentile(all_grads, 25, axis=0), np.percentile(all_grads, 75, axis=0)

def eval_and_plot(grad_dict: dict, alpha, marker, linewidth, linestyle, label, color=None, ax=None):
    if ax is None:
        ax = plt.gca()
    keys = list(grad_dict.keys())
    values = np.array(list(grad_dict.values()))
    medians, lower_percentile, upper_percentile = bootstrap_gradient_variance(values)
    ax.plot(keys, medians, marker=marker, linewidth=linewidth, linestyle=linestyle, label=label, ms=11, color=color, alpha=0.9, rasterized=True)
    ax.fill_between(keys, upper_percentile, lower_percentile, alpha=alpha, color=color, rasterized=True)
```
